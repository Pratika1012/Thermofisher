```python
import fitz  # PyMuPDF
from PIL import Image
import io
import os
import re

def extract_image_titles_from_page(page):
    titles = []
    text = page.get_text("text")
    for line in text.split('\n'):
        match = re.match(r"Figure \d+\. .+", line)
        if match:
            titles.append(match.group(0))
    return titles

def extract_images_from_pdf(pdf_path, output_folder):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    image_count = 0

    # Create the output directory if it does not exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)
        image_list = page.get_images(full=True)
        titles = extract_image_titles_from_page(page)
        
        for img_index, img in enumerate(image_list):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image = Image.open(io.BytesIO(image_bytes))
            
            # Convert CMYK to RGB if needed
            if image.mode == "CMYK":
                image = image.convert("RGB")
            
            # Use the title or a default name if title is not available
            title = titles[img_index] if img_index < len(titles) else f"Figure_{page_number + 1}_{img_index + 1}"
            title = title.replace(':', '')  # Remove invalid characters for filenames
            image_filename = os.path.join(output_folder, f"{title}.png")
            image.save(image_filename)
            image_count += 1

    return image_count

def main():
    pdf_path = r"Data Input Document 4_User Manual.pdf"
    output_folder = "ExtractedImages4"
    image_count = extract_images_from_pdf(pdf_path, output_folder)
    print(f"Extracted {image_count} images and saved in {output_folder}.")

if __name__ == "__main__":
    main()
```

```python
import os
import re
import io
from docx import Document
from PIL import Image

def extract_image_titles_from_paragraphs(paragraphs):
    titles = []
    for para in paragraphs:
        match = re.match(r"Figure \d+\. .+", para.text)
        if match:
            titles.append(match.group(0))
    return titles

def extract_images_from_docx(docx_path, output_folder):
    # Open the Word document
    doc = Document(docx_path)
    image_count = 0

    # Create output folder if it does not exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Extract figure titles from paragraphs
    titles = extract_image_titles_from_paragraphs(doc.paragraphs)

    # Extract images and save them
    for rel in doc.part.rels.values():
        if "image" in rel.target_ref:  # Check if it's an image relationship
            image_data = rel.target_part.blob
            img = Image.open(io.BytesIO(image_data))

            # Use the title or a default name if title is not available
            title = titles[image_count] if image_count < len(titles) else f"Figure_{image_count + 1}"
            title = title.replace(':', '')  # Remove invalid characters for filenames
            image_filename = os.path.join(output_folder, f"{title}.png")
            
            # Save the image
            img.save(image_filename)
            image_count += 1

    return image_count

def main():
    docx_path = r"Input File - Rework Procedure (1).docx"
    output_folder = "ExtractedImagesWord"
    image_count = extract_images_from_docx(docx_path, output_folder)
    print(f"Extracted {image_count} images and saved in {output_folder}.")

if __name__ == "__main__":
    main()
```

```python
import pdfplumber
import pandas as pd
import re
import os

def extract_table_titles_from_page(page):
    """
    Extract potential table titles from the page text.
    Looks for patterns like 'Table 1. Some description'.
    """
    titles = []
    text = page.extract_text()
    for line in text.split('\n'):
        match = re.match(r"(Table \d+\. .+)", line)
        if match:
            titles.append(match.group(0))
    return titles

def extract_and_format_tables(pdf_path):
    def clean_sheet_name(name):
        # Remove invalid characters and trim the length to be Excel compliant
        name = re.sub(r'[\\/*?:[\]]', '', name)  # Remove invalid characters
        return name[:31]  # Excel sheet names are limited to 31 characters

    # Step 1: Load the PDF and extract tables
    with pdfplumber.open(pdf_path) as pdf:
        all_tables = []
        sheet_names = []

        for page_number, page in enumerate(pdf.pages):
            tables = page.extract_tables()
            titles = extract_table_titles_from_page(page)  # Extract possible titles from the page
            
            for table_index, table in enumerate(tables):
                if not table:
                    continue  # Skip if the table is empty

                # Reconstruct each table, placing content properly
                max_columns = max(len(row) for row in table)
                formatted_table = []
                for row in table:
                    formatted_row = [cell.strip() if cell else '' for cell in row]
                    while len(formatted_row) < max_columns:
                        formatted_row.append('')
                    formatted_table.append(formatted_row)

                # Use the title if available, otherwise use a fallback name
                if table_index < len(titles):
                    sheet_name = clean_sheet_name(titles[table_index])
                else:
                    sheet_name = f"Table_{page_number + 1}_{table_index + 1}"  # Fallback name

                # Convert the table to a DataFrame
                df = pd.DataFrame(formatted_table[1:], columns=formatted_table[0])  # Skip the first row as it is the header
                all_tables.append(df)
                sheet_names.append(sheet_name)
    
    return all_tables, sheet_names

def save_tables_to_excel(tables, sheet_names, output_path):
    # Step 2: Save each table as a sheet in an Excel file
    with pd.ExcelWriter(output_path) as writer:
        for table, sheet_name in zip(tables, sheet_names):
            table.to_excel(writer, sheet_name=sheet_name, index=False)

# Example usage:
pdf_path = "Data Input Document 1_User Manual.pdf"
output_path = "extracted_tables.xlsx"

# Extract and format tables
tables, sheet_names = extract_and_format_tables(pdf_path)

# Save the formatted tables to an Excel file
save_tables_to_excel(tables, sheet_names, output_path)

print(f"Tables saved to {output_path}")
```

```python
import os
import io
import subprocess
from PIL import Image
import fitz  # PyMuPDF for handling PDFs
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTImage, LTFigure
from docx2pdf import convert  # For converting .docx files to PDF

# Path to the directory containing PDF and Word files
input_folder = 'Input File - Rework Procedure (1).docx'
output_folder = 'extracted_images_CER_2'

# Create the output folder if it does not exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Function to render a full PDF page as an image
def render_full_page_as_image(pdf_path, page_number, output_folder, base_filename):
    try:
        pdf_file = fitz.open(pdf_path)
        page = pdf_file.load_page(page_number)
        zoom = 2  # Increase this for higher resolution
        mat = fitz.Matrix(zoom, zoom)
        pix = page.get_pixmap(matrix=mat)

        # Define the path for the saved image
        full_page_image_path = os.path.join(output_folder, f"{base_filename}_page_{page_number + 1}.png")

        # Save the rendered page as an image
        pix.save(full_page_image_path)
        print(f"Rendered full page {page_number + 1} as an image to {full_page_image_path}")

        pdf_file.close()
    except Exception as e:
        print(f"Failed to render full page {page_number + 1} of {base_filename}. Reason: {e}")

# Function to extract vector graphics and handle unsupported images by rendering full pages
def extract_vector_graphics(pdf_path, output_folder, base_filename):
    vector_image_counter = 1  # Counter to ensure unique names for vector images
    try:
        for page_number, page_layout in enumerate(extract_pages(pdf_path)):
            full_page_rendered = False  # Flag to check if full page render is done
            for element in page_layout:
                if isinstance(element, LTFigure):
                    figure_counter = 1  # Reset counter for each figure within a page
                    for fig in element:
                        if isinstance(fig, LTImage):
                            try:
                                # Get image data
                                image_data = fig.stream.get_data()

                                # Check the first few bytes to determine type
                                header = image_data[:10]
                                print(f"Header bytes for {base_filename}, page {page_number + 1}: {header}")

                                # Check for common headers
                                if header.startswith(b'\xff\xd8'):  # JPEG
                                    image = Image.open(io.BytesIO(image_data))
                                    image.verify()  # Verify if the image is not corrupted
                                    image = Image.open(io.BytesIO(image_data))  # Reopen after verification
                                    image = image.convert('RGB')  # Convert to standard format

                                    # Ensure unique filename for each vector graphic
                                    image_path = os.path.join(output_folder, f"{base_filename}_vector_page_{page_number + 1}_{vector_image_counter}_{figure_counter}.png")
                                    image.save(image_path)
                                    print(f"Saved vector graphic from {base_filename}, page {page_number + 1} to {image_path}")

                                    # Increment counters
                                    vector_image_counter += 1
                                    figure_counter += 1
                                else:
                                    # Handle non-standard or unsupported headers by rendering the full page
                                    if not full_page_rendered:
                                        render_full_page_as_image(pdf_path, page_number, output_folder, base_filename)
                                        full_page_rendered = True

                                    # Increment counters
                                    vector_image_counter += 1
                                    figure_counter += 1
                            except (IOError, SyntaxError) as e:
                                print(f"Error saving vector graphic on page {page_number + 1} of {base_filename}: {e}")
                            except Exception as ex:
                                # If it fails, print and skip this item
                                print(f"Unknown format or error at page {page_number + 1} of {base_filename}: {ex}")
    except Exception as e:
        print(f"Failed to extract vector graphics from {base_filename}. Reason: {e}")

# Function to convert .doc files to PDF using LibreOffice
def convert_doc_to_pdf(input_path, output_folder):
    try:
        subprocess.run(['libreoffice', '--headless', '--convert-to', 'pdf', input_path, '--outdir', output_folder], check=True)
        print(f"Converted {input_path} to PDF in {output_folder}")
    except subprocess.CalledProcessError as e:
        print(f"Failed to convert {input_path} to PDF. Reason: {e}")

# Function to process all PDF and Word files in the input folder
def process_files(input_folder, output_folder):
    for filename in os.listdir(input_folder):
        file_path = os.path.join(input_folder, filename)
        base_filename = os.path.splitext(filename)[0]

        # Check if the file is a PDF
        if filename.lower().endswith('.pdf'):
            print(f"Processing PDF file: {filename}")
            extract_vector_graphics(file_path, output_folder, base_filename)

        # Check if the file is a Word document (.docx)
        elif filename.lower().endswith('.docx'):
            print(f"Processing Word file: {filename}")
            # Convert the Word document to PDF
            pdf_converted_path = os.path.join(input_folder, f"{base_filename}.pdf")
            try:
                convert(file_path, pdf_converted_path)
                print(f"Converted {filename} to PDF: {pdf_converted_path}")

                # Process the converted PDF
                extract_vector_graphics(pdf_converted_path, output_folder, base_filename)
            except Exception as e:
                print(f"Failed to convert {filename} to PDF. Reason: {e}")

        # Check if the file is a Word document (.doc)
        elif filename.lower().endswith('.doc'):
            print(f"Processing Word file: {filename}")
            # Convert the .doc document to PDF
            try:
                convert_doc_to_pdf(file_path, output_folder)
                pdf_converted_path = os.path.join(output_folder, f"{base_filename}.pdf")

                # Process the converted PDF
                extract_vector_graphics(pdf_converted_path, output_folder, base_filename)
            except Exception as e:
                print(f"Failed to convert {filename} to PDF. Reason: {e}")

# Run the function to process files in the input directory
process_files(input_folder, output_folder)
```

```python
import re
from PyPDF2 import PdfReader
from docx import Document

def extract_text_with_pyPDF(pdf_file_path, output_docx_path):
    # Read the PDF
    pdf_reader = PdfReader(pdf_file_path)
    
    # Initialize a Word document
    doc = Document()
    
    # Regex for detecting page numbers (e.g., "Page X", "X of Y", simple numbers)
    page_number_pattern = re.compile(r'(^\d+$)|(^Page\s*\d+)|(\d+\s+of\s+\d+)', re.IGNORECASE)

    for page in pdf_reader.pages:
        text = page.extract_text()
        
        if text:
            # Collect lines of text from the page
            lines = [line.strip() for line in text.split('\n') if line.strip()]
            
            # Remove the footer: assume the footer is the last line or last few lines on the page
            # Remove last 1-2 lines if they match the footer or page number patterns
            if len(lines) > 1:
                last_line = lines[-1]
                
                # Check if the last line looks like a footer or page number
                if page_number_pattern.match(last_line) or len(last_line.split()) <= 3:
                    lines = lines[:-1]  # Remove the last line
                
                # Optional: you can check if the second-to-last line is also a footer (for multi-line footers)
                if len(lines) > 1 and (page_number_pattern.match(lines[-1]) or len(lines[-1].split()) <= 3):
                    lines = lines[:-1]  # Remove the second-to-last line as well
            
            # Now write the remaining filtered lines (without the footer) to the Word document
            for line in lines:
                doc.add_paragraph(line)
    
    # Save the extracted and filtered text to a Word file
    doc.save(output_docx_path)

# Example usage:
pdf_file_path = "Data Input Document 1_User Manual.pdf"
output_docx_path = "Data Input Document 1_User Manual.docx"
extract_text_with_pyPDF(pdf_file_path, output_docx_path)
```

```python
import fitz  # PyMuPDF
from docx import Document
import re

def clean_text(text):
    """
    Remove control characters and invalid XML characters from the text.
    """
    # Remove non-XML-compatible characters
    cleaned_text = re.sub(r'[^\x20-\x7E\n\r\t]', '', text)
    return cleaned_text

def extract_pdf_text(pdf_path):
    """
    Extract text from the PDF while ignoring the footer and cleaning the text.
    """
    doc = fitz.open(pdf_path)
    extracted_text = []

    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        page_height = page.rect.height
        blocks = page.get_text("blocks")  # Extract text blocks with positions
        cleaned_lines = []

        for block in blocks:
            block_text = block[4]  # Text content of the block
            block_y0 = block[1]    # Y-coordinate (top) of the block
            block_y1 = block[3]    # Y-coordinate (bottom) of the block

            # Skip text near the bottom of the page (likely footers)
            if block_y0 > page_height * 0.85 or block_y1 > page_height * 0.85:
                continue

            # Clean text and add to the lines
            cleaned_lines.append(clean_text(block_text))

        # Combine cleaned lines for the page
        page_text = "\n".join(cleaned_lines)
        extracted_text.append(page_text)

    doc.close()
    return extracted_text

def save_to_word(text_list, output_path):
    """
    Save the extracted text into a Word document.
    """
    document = Document()
    
    for page_text in text_list:
        document.add_paragraph(page_text)
    
    document.save(output_path)

# Path to the uploaded PDF file
pdf_path = 'Data Input Document 2_User Manual.pdf'
word_output_path = 'Data Input Document 2_User Manual.docx'

# Extract text from PDF and save to Word
extracted_text = extract_pdf_text(pdf_path)
save_to_word(extracted_text, word_output_path)

print(f"Extracted text has been saved to {word_output_path}")
```

```python
import fitz  # PyMuPDF for PDF processing
import pytesseract
from PIL import Image
import io
from PyPDF2 import PdfReader
from docx import Document
import google.generativeai as genai  # Assuming this is the library for Gemini Model

# Configure the API key for the Gemini Model
genai.configure(api_key="AIzaSyBNKJ5UoqldD8BwNVCwbDHs3GquaZ9OIjM")

# Set up the model configuration for text generation
generation_config = {
    "temperature": 0.8,            # Controls randomness in output
    "top_p": 0.9,                  # Controls nucleus sampling for creative responses
    "top_k": 50,                   # Limits the highest-probability tokens considered
    "max_output_tokens": 2048,     # Max number of tokens allowed for generation
}

# Safety settings to ensure that generated content is safe
safety_settings = [
    {"category": "HARM_CATEGORY_HARASSMENT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
    {"category": "HARM_CATEGORY_HATE_SPEECH", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
    {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
    {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"}
]

# Initialize the Gemini model
model = genai.GenerativeModel(
    model_name="gemini-pro",         # Model name
    generation_config=generation_config,
    safety_settings=safety_settings
)

# Function to extract images from the PDF
import os  # Import os to handle directory operations

def extract_images_from_pdf(pdf_file_path, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    pdf_document = fitz.open(pdf_file_path)
    images = []
    
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        image_list = page.get_images(full=True)
        
        for image_index, img in enumerate(image_list):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            image_filename = f"{output_folder}/control_panel_image_{page_num + 1}_{image_index + 1}.{image_ext}"
            
            # Save the image
            with open(image_filename, "wb") as image_file:
                image_file.write(image_bytes)
            
            images.append(image_filename)
    
    return images


# Function to extract text from PDF
def extract_text_from_pdf(pdf_file_path):
    reader = PdfReader(pdf_file_path)
    text = ''
    
    for page in reader.pages:
        text += page.extract_text() + '\n'
    
    return text

# Function to perform OCR on images
def ocr_image(image_path):
    with Image.open(image_path) as img:
        text = pytesseract.image_to_string(img)
    return text

# Function to generate the Word document based on the template
def generate_word_document(images, description, output_docx_path, gemini_response):
    doc = Document()

    # Add the title for the Control Panel
    doc.add_heading('Control Panel', level=1)

    # Add each image with its description
    for image_path in images:
        doc.add_paragraph(f"Image: {image_path}")
        doc.add_picture(image_path)
        doc.add_paragraph()

    # Add the extracted description
    doc.add_paragraph("Extracted Description:")
    doc.add_paragraph(description)

    # Add the Gemini Model's response
    doc.add_paragraph("\nGemini Model Response:")
    doc.add_paragraph(gemini_response)

    # Save the document
    doc.save(output_docx_path)

# Function to generate response from Gemini Model
def generate_gemini_response(description):
    prompt = f"Generate a detailed description based on the following control panel information:\n\n{description}"
    
    # Generate response from the model
    response = model.generate_content(prompt=prompt)
    
    # Extract generated text
    if 'output' in response:
        return response['output']
    else:
        return "No valid response generated."

# Function to process the PDF and generate the response
def process_pdf_and_generate_output(pdf_file_path, output_folder, output_docx_path):
    # Step 1: Extract control panel images
    images = extract_images_from_pdf(pdf_file_path, output_folder)

    # Step 2: Extract control panel-related text
    raw_text = extract_text_from_pdf(pdf_file_path)

    # Step 3: Locate the "Control Panel" section and relevant description
    control_panel_start = raw_text.lower().find('control panel')
    if control_panel_start != -1:
        control_panel_description = raw_text[control_panel_start: control_panel_start + 1500]  # Extract up to 1500 chars around control panel
    else:
        control_panel_description = "No control panel description found."

    # Step 4: Generate the Gemini Model response
    gemini_response = generate_gemini_response(control_panel_description)

    # Step 5: Perform OCR on each image and extract descriptions
    image_descriptions = {}
    for image_path in images:
        image_text = ocr_image(image_path)
        image_descriptions[image_path] = image_text

    # Use the descriptions from images to further refine or modify the document
    detailed_description = "\n".join([f"{img}: {desc}" for img, desc in image_descriptions.items()])

    # Step 6: Generate the Word document with images and descriptions
    generate_word_document(images, detailed_description, output_docx_path, gemini_response)
    print(f"Generated response saved to {output_docx_path}")

# Example usage
pdf_file_path = "Data Input Document 1_User Manual.pdf"  # Path to your PDF file
output_folder = "./images"  # Folder to save extracted images
output_docx_path = "Generated_Response_Control_Panel.docx"  # Path for the output Word document

# Run the main function
process_pdf_and_generate_output(pdf_file_path, output_folder, output_docx_path)
```
